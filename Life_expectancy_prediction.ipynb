**4. Impute the missing values**

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean',fill_value=None)
df['Life expectancy ']=imputer.fit_transform(df[['Life expectancy ']])
df['Adult Mortality']=imputer.fit_transform(df[['Adult Mortality']])
df['Alcohol']=imputer.fit_transform(df[['Alcohol']])
df['Hepatitis B']=imputer.fit_transform(df[['Hepatitis B']])
df[' BMI ']=imputer.fit_transform(df[[' BMI ']])
df['Polio']=imputer.fit_transform(df[['Polio']])
df['Total expenditure']=imputer.fit_transform(df[['Total expenditure']])
df['Diphtheria ']=imputer.fit_transform(df[['Diphtheria ']])
df['GDP']=imputer.fit_transform(df[['GDP']])
df['Population']=imputer.fit_transform(df[['Population']])
df[' thinness  1-19 years']=imputer.fit_transform(df[[' thinness  1-19 years']])
df[' thinness 5-9 years']=imputer.fit_transform(df[[' thinness 5-9 years']])
df['Income composition of resources']=imputer.fit_transform(df[['Income composition of resources']])
df['Schooling']=imputer.fit_transform(df[['Schooling']])
```

```python
#checking if there is any null values left
df.isnull().sum()
```

 **5. Find the count, mean, standard deviation, quartiles**

```python
df.describe()
```

**6. Correlation Heat Map**

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_df = df[numeric_columns]
corr_matrix = numeric_df.corr()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(corr_matrix, cmap = 'coolwarm', vmin = -1, vmax = 1, center = 0, annot = True)
plt.show()
```

There are strong correlation between:

*  infant deaths and under-five deaths
*  GPD and percentage expenditure
* thinness 1-19 years and thinness 5-9 years

**8. Comparation the life expectancy of the developed countries to developing countries by using violin plots.**

**9. Line Plot**

```python
Can_df = df[df["Country"] == 'Canada']
UK_df = df[df["Country"] == 'United Kingdom of Great Britain and Northern Ireland']
US_df = df[df["Country"] == 'United States of America']
```

```python
countries_3 = [Can_df, UK_df, US_df]
```

```python
df_countries = pd.concat(countries_3)
```

```python
fig=px.line(df_countries.sort_values(by='Year'),x='Year',y='Life expectancy ',color='Country', markers=True,template='plotly_dark',title='<b> Country wise Life Expectancy over Years')
fig.show()
```

The Life Expectancy of Canada is generally higher compared to United States of America and United Kingdom of Great Britain and Northern Ireland.

**10. The average infant deaths over the years against the average life expectancy**

```python
Bel_df = df[df["Country"] == 'Belgium']
Brazil_df = df[df["Country"] == 'Brazil']
Cameroon_df = df[df["Country"] == 'Cameroon']
CN_df = df[df["Country"] == 'China']
Fr_df = df[df["Country"] == 'France']
Gha_df = df[df["Country"] == 'Ghana']
IN_df = df[df["Country"] == 'India']
```

```python
countries_7 = [Can_df, UK_df, US_df, Bel_df, Brazil_df, Cameroon_df, CN_df, Fr_df, Gha_df, IN_df]
```

```python
df_7countries = pd.concat(countries_7)
```

```python
px.scatter(df_7countries.sort_values(by='Year'),y='infant deaths',x='Life expectancy ',template='plotly_dark',size='Year',color='Country',opacity=0.6,title='<b>Life Expectancy Versus Infant Deaths of Countries in every Year')
```

The group of developed countries such as France, Belgium, Canada, United States of America and United Kingdom of Great Britain and Northern Ireland have longer Life expectancy along with minimal infant deaths. The developing countries such as Ghana, Cameroon, China, India and Brazil, either have short Life expectancy or high infant deaths average.
